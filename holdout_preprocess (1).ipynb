```python
import pandas as pd
import pycountry 
import pycountry_convert as pc 
import numpy as np
```

```python
oceancrest_df = pd.read_csv('OceanCrest_holdout_noresponse.csv')
```

```python
oceancrest_df.columns
```

```python
#oceancrest_df.drop('Unnamed: 0', axis=1, inplace=True)
oceancrest_df.head()
```

```python
oceancrest_df['is_agent'] = np.where(oceancrest_df['agent'].isna(), 0, 1)
```

```python
oceancrest_df['is_company'] = np.where(oceancrest_df['company'].isna(), 0, 1)
```

```python
mask = oceancrest_df['reserved_room_type'] == oceancrest_df['assigned_room_type']
oceancrest_df['is_reserved_room'] = np.where(mask, "reserved", "notreserved")
```

```python
oceancrest_df.isna().sum()
```

```python

oceancrest_df.country = np.where(oceancrest_df.country == 'CN', 'CHN', oceancrest_df['country'])
oceancrest_df.country.unique()
```

```python
def alpha3_to_continent(alpha_3):
    try:
        # Convert from alpha-3 to alpha-2
        alpha_2 = pycountry.countries.get(alpha_3=alpha_3).alpha_2
        # Get continent code from alpha-2 country code
        continent_code = pc.country_alpha2_to_continent_code(alpha_2)
        # Convert continent code to continent name
        return pc.convert_continent_code_to_continent_name(continent_code)
    except (AttributeError, KeyError):
        return None


oceancrest_df['country'] = oceancrest_df['country'].astype(str)
oceancrest_df['continent'] = oceancrest_df['country'].apply(alpha3_to_continent)
```

```python
filter1 = oceancrest_df.country == 'TMP'
filter2 = oceancrest_df.country == 'UMI'
filter3 = oceancrest_df.country == 'ATA'
filter4 = oceancrest_df.country == 'ATF'
oceancrest_df.loc[filter1, 'continent'] = 'Asia'
oceancrest_df.loc[filter2, 'continent'] = 'North America'
oceancrest_df.loc[filter3, 'continent'] = 'Antarctica'
oceancrest_df.loc[filter4, 'continent'] = 'Antarctica'
# array(['nan', 'TMP', 'UMI', 'ATA', 'ATF'], dtype=object)
```

```python
oceancrest_df[['country', 'continent']]
```

```python
mask2 = oceancrest_df['meal'] == 'Undefined'

oceancrest_df['is_meal'] = np.where(mask2, 0, np.where(oceancrest_df['meal'].isna(), np.nan, 1))
```

```python
oceancrest_df[['meal', 'is_meal']]
```

```python
season_mapping = {
    'Spring': ['March', 'April', 'May'],
    'Summer': ['June', 'July', 'August'],
    'Fall': ['September', 'October', 'November'],
    'Winter': ['December', 'January', 'February']
}

month_to_season = {month: season for season, months in season_mapping.items() for month in months}
oceancrest_df['arrival_season'] = oceancrest_df.arrival_date_month.map(month_to_season)
```

```python
oceancrest_df[['arrival_season', 'arrival_date_month']]
```

```python
oceancrest_df['total_length_of_stay'] = oceancrest_df.stays_in_week_nights + oceancrest_df.stays_in_weekend_nights
```

```python
oceancrest_df['total_booking'] = oceancrest_df.previous_bookings_not_canceled + oceancrest_df.previous_cancellations
```

```python
oceancrest_df.isna().sum()
```

```python

oceancrest_df.drop(['arrival_date_month', 'arrival_date_week_number', 'arrival_date_day_of_month', 'meal', 'country', 'reserved_room_type', 'assigned_room_type', 'company', 'agent', 'reservation_status', 'reservation_status_date'], axis=1, inplace=True)
oceancrest_df.head()
```

```python
# Replace missing values in the 'Age' column with the mode
mode_age = oceancrest_df['children'].mode()[0]
oceancrest_df['children'].fillna(mode_age, inplace=True)
```
